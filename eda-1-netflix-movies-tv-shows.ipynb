**Step 1: Select a real-world dataset**

```python
import pandas as pd  #Import Libraries
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")  # Load dataset
```

```python
df.head()  # Display first 5 rows
```

**Step 2: Perform data preparation & cleaning**

```python
df.shape  # Shape of dataset (rows, cols)
```

```python
df.info()  # Info about datatypes & missing values
```

```python
df.isnull().sum()  # Missing values count
```

```python
df.describe()  # Summary statistics for numeric columns
```

```python
df['country'].fillna("Unknown",inplace=True)  # Fill missing 'country' with "Unknown"
df['date_added'].fillna(df['date_added'].mode()[0],inplace=True)  # Fill missing 'date_added' with mode
df['director'].fillna("Unknown",inplace=True)  # Fill missing 'directo' with "Unknown"
df['cast'].fillna("Unknown",inplace=True)  # Fill missing 'cast' with "Unknown"
```

```python
df.dropna(subset=['rating'], inplace=True)  # Drop rows with missing rating info
df.dropna(subset=['duration'], inplace=True)  # Drop rows with missing duration info
```

```python
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)  # Fill missing 'date_added' with mode
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])  # Convert date to datetime
```

**Step 3: Perform exploratory analysis & visualization Matplotlib - Resources**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="type", data=df, palette="Set2")  # Count of Movies vs TV Shows
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)  # Top 10 countries producing Netflix content
top_countries.plot(kind="bar", color="teal")
plt.title("Top 10 Countries by Number of Netflix Titles")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))  # Distribution of Release Years
df['release_year'].hist(bins=30, color="orange")
plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year  # Content added trend over years
df['year_added'].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Trend of Content Added Over Years")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.show()
```

**Step 4: Ask & answer questions about the data**

**Q1. Which country has produced the most Netflix content?**

```python
df['country'].value_counts().head(1)
```

**Q2. What’s the most common genre (listed in “listed_in” column)?**

```python
from collections import Counter
genres = []
for i in df['listed_in']:
    genres.extend(i.split(", "))
Counter(genres).most_common(5)
```

**Q3. How many titles were released each decade?**

```python
df['decade'] = (df['release_year']//10)*10
df['decade'].value_counts().sort_index()
```

**Q4. Are movies generally longer than TV Shows (by duration)?**

```python
df_movies = df[df['type']=="Movie"].copy()
df_movies['duration'] = df_movies['duration'].str.replace(" min","").astype(float)

print("Average Movie Duration:", df_movies['duration'].mean(), "minutes")
```
